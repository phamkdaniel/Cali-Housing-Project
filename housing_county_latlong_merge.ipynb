```python
import pandas as pd
import numpy as np
```

```python
file = "Resources/housing.csv"
housing_df = pd.read_csv(file)
housing_df.head()
```

```python
round_long = housing_df["longitude"].round(1)
round_long
```

```python
round_lat = housing_df["latitude"].round(1)
round_lat
```

```python
housing_df.insert(10,"longitude_round",round_long,True)
housing_df.insert(11,"latitude_round",round_lat,True)
housing_df
```

```python
county_file = "Resources/county_data_clean.csv"
county_df = pd.read_csv(county_file)
county_df = pd.DataFrame(county_df)
county_df = county_df[["latitude", "longitude","county_name"]]

county_df
```

```python
housing_county_df = pd.merge(housing_df, county_df, how="left", left_on=["longitude_round","latitude_round"], right_on=["longitude","latitude"])

housing_county_df.head()
```

```python
housing_county_df.to_csv("Resources/housing_county_df.csv")
```

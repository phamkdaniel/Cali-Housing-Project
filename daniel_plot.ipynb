```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import csv
housing_data_df = pd.read_csv("Resources/housing.csv")
housing_data_df
```

```python
# create fig, axes
fig, ax = plt.subplots(figsize=(10,10))

# force equal aspect ratio
ax.set_aspect('equal')

# get coordinate
lat = housing_data_df["latitude"]
long = housing_data_df["longitude"]

# set axes limits
ax.set_xlim([min(long), max(long)])
ax.set_ylim([min(lat), max(lat)])

# get coloration for house value
coloring = housing_data_df.median_house_value

# create scatterplot
house_val = ax.scatter(long, lat, c=coloring, cmap='viridis', alpha=0.2)

# create x,y labels
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

# set axes for colorbar
caxes = fig.add_axes([0.95, 0.1, 0.03, 0.8]) 
cbar = fig.colorbar(house_val, cax=caxes, orientation='vertical')
cbar.set_label("Median House Value", fontsize=14)

# overlay outline of California
xlim = ax.get_xlim()
ylim = ax.get_ylim()
im = plt.imread("Images/cali_outline.png")
implot = ax.imshow(im, extent=[xlim[0], xlim[1], ylim[0], ylim[1]])

# save image
# plt.savefig("Images/cali_housing.png")

# show plot
plt.show()
```
